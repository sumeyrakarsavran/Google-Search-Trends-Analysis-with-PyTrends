```python
from pytrends.request import TrendReq #paket çok büyük old için bir kısmını kullanıyoruz
pytrends=TrendReq()

keywords=['Python','R','C++','Java','HTML']
pytrends.build_payload(keywords,timeframe='today 5-y')
```

```python
df=pytrends.interest_over_time()
del df['isPartial'] #o sütun gereksiz diye silelim
df.head()
```

```python
df.plot()
```
